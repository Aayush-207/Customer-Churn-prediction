```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = {
    "Logistic Regression": {
        "Accuracy": 0.705465,
        "Precision": 0.470073,
        "Recall": 0.860963,
        "F1 Score": 0.608121,
        "ROC-AUC": 0.838229
    },
    "Random Forest": {
        "Accuracy": 0.770759,
        "Precision": 0.549708,
        "Recall": 0.754011,
        "F1 Score": 0.635851,
        "ROC-AUC": 0.844190
    },
    "XGBoost": {
        "Accuracy": 0.777147,
        "Precision": 0.560241,
        "Recall": 0.745989,
        "F1 Score": 0.639908,
        "ROC-AUC": 0.847787
    },
    "LightGBM": {
        "Accuracy": 0.781405,
        "Precision": 0.571121,
        "Recall": 0.708556,
        "F1 Score": 0.632458,
        "ROC-AUC": 0.826433
    }
}
```

```python
results_df = pd.DataFrame(results).T
results_df
```

```python
best_models = {}

for metric in results_df.columns:
    best_models[metric] = results_df[metric].idxmax()

best_models
```

```python
plt.figure()
plt.bar(results_df.index, results_df["Accuracy"])
plt.title("Model Comparison - Accuracy")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
plt.bar(results_df.index, results_df["Precision"])
plt.title("Model Comparison - Precision")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
plt.bar(results_df.index, results_df["Recall"])
plt.title("Model Comparison - Recall")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
plt.bar(results_df.index, results_df["F1 Score"])
plt.title("Model Comparison - F1 Score")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
plt.bar(results_df.index, results_df["ROC-AUC"])
plt.title("Model Comparison - ROC-AUC")
plt.xticks(rotation=45)
plt.show()
```

```python
print(" Best Model Per Metric:\n")

for metric, model in best_models.items():
    print(f"{metric}: {model}")
```
